```python
import torch 
import random
import matplotlib 
import pandas as pd
import seaborn as sns
```

```python
inputs = []

for _ in range(100_000):
    new_num = random.random()*100*random.choice([-1,1])
    inputs.append(new_num)
```

```python
outputs = [x**2+random.random()*4*random.choice([-1,1]) for x in inputs]
```

```python
raw_data = pd.DataFrame({'X':inputs,'Y':outputs})
```

```python
sns.pairplot(raw_data)
```
